## <><b>Test Stationarity</b><p>
A stationary time series is one that its statistical properties are stable over time. In order to check if the time series is stationary, I will use the Augmented Dickey-Fuller (ADF) test.

The ADF is a hypothesis test which the null hypothesis is that the time series is no-stationary, if we can reject the null hypothesis then our time series is stationary. For this I will use the statsmodels library adfuller method.


***The data for the variable taken is not stationary, it is showing upward trend. So, Augmented Dickey Fuller test will be conducted to check for unit root.***

````python
from statsmodels.tsa.stattools import adfuller
results=adfuller(mk['Close'],autolag='AIC')
print('ADF Statistics: %f' %results[0])
print('P-value: %f' % results[1])
````

<h> Here we can see that our p value is greater than 0.05 though our data is no stationary so for applying Arima Model we haave to convert our data in stationary <p> we will use log difference method for stationary data</h>

````python
mks=np.log(mk['Close']).diff() # difference with log-1 and log value for minimize fluctuation
mks.head()
````

````python
mks.dropna(inplace=True)# remove the nan value 
````

````python
mks
````

<h> we will check stationarity again if it stationary then we will use arima model otherwise we will diff. again</h>

````python
from statsmodels.tsa.stattools import adfuller
results=adfuller(mks,autolag='AIC')
print('ADF Statistics: %f' %results[0])
print('P-value: %f' % results[1])
````

````python
# here our p value is 0 or less than 0.05 so our data is stationary as showen in graph
plt.plot(mks)
````

# ***ARIMA MODEL***

***What is ARIMA model?***<p>
Autoregressive integrated moving average (ARIMA) models predict future values based on past values. ARIMA makes use of lagged moving averages to smooth time series data. They are widely used in technical analysis to forecast future security prices.

````python
from statsmodels.tsa.arima_model import ARIMA # imported ARIMA from statsmodel
model_arima=ARIMA(train,order=(4,1,0))
model_arima_fit=model_arima.fit()
````

````python
forecast=model_arima_fit.forecast(steps=20)[0] # we have forecasted the value for 20 steps
forecast
````

````python
plt.plot(test,label='actual value')
plt.plot(forecast,color='pink',label='forecasted value')
plt.legend()
````
